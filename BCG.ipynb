**5. Discussion**

The line plot describes the data for BCG vaccination of the percentage of the persons vaccinated in the years '1997','2002','2007', '2012','2017'. This visualization answers the following thoughts:
1. Almost most of the countries have achieved near 80% - 100% vaccination of their population
2. But, the situation looks grim in certain countries which have poor vaccination rates compared to many countries.
3. Countries (less than 80% population covered) with that have not achieved maximum vaccination cover include Bhutan, DRC, CostaRica, Madagascar, Pakistan, Somalia etc.
4. It is astnonishing to find that less than 20% of the Sweden's population is vaccinated.

The plot also describes the the general pattern of increase in the number of people covered in all the partner countries. There is a widespread increase in the number of people vaccinted globally when compared between 1997 and 2017. This is a positive sign for the policies and efforts of the UNICEF's vaccination programs. 

Special considerations need to be takien for the reduced population being vaccinated in Sweden, because the most of the countries which are yet to achieve 80% vaccination cover include countries in the East, i.e, Asian and African countries. It is very surprising to find Sweden below 20% vaccination cover.
